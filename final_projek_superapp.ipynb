```python
# menjumlahkan yang terindikasi outliers
Q1 = df[list_numerik_features].quantile(0.25)
Q3 = df[list_numerik_features].quantile(0.75)
IQR = Q3 - Q1
outliers_count_specified = ((df[list_numerik_features] < (Q1 - 1.5 * IQR)) | (df[list_numerik_features] > (Q3 + 1.5 * IQR))).sum()

outliers_count_specified
```

```python
# Membuat boxplot untuk setiap fitur numerik
plt.figure(figsize=(10, 6))
for i, feature in enumerate(list_numerik_features):
    plt.subplot(3, 2, i + 1)
    sns.boxplot(x=df[feature], color='skyblue')
    plt.title(f'Boxplot of {feature}')
    plt.xlabel(feature)

plt.tight_layout()
plt.show()
```

```python
# handling outliers
def handle_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Menentukan batas bawah dan batas atas
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Mengganti nilai outlier dengan nilai batas atas atau batas bawah terdekat
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])

# Menggunakan fungsi handle_outliers_iqr untuk setiap kolom yang memiliki outlier
list_numerik_features
for column in list_numerik_features:
    handle_outliers_iqr(df, column)
```

```python
# Membuat boxplot untuk setiap fitur numerik
plt.figure(figsize=(10, 6))
for i, feature in enumerate(list_numerik_features):
    plt.subplot(3, 2, i + 1)
    sns.boxplot(x=df[feature], color='skyblue')
    plt.title(f'Boxplot of {feature}')
    plt.xlabel(feature)

plt.tight_layout()
plt.show()
```

```python
df_september = df[df['transaction_date'].dt.month == 9]
df_october = df[df['transaction_date'].dt.month == 10]

# Menghitung jumlah transaksi harian untuk bulan September
daily_transactions_september = df_september.groupby(df_september['transaction_date'].dt.day).size()

# Menghitung jumlah transaksi harian untuk bulan Oktober
daily_transactions_october = df_october.groupby(df_october['transaction_date'].dt.day).size()
```

```python
# Plot transaksi harian untuk bulan September
plt.figure(figsize=(12, 6))
plt.plot(daily_transactions_september.index, daily_transactions_september.values, marker='o', label='September')
plt.title('Daily Transactions in September')
plt.xlabel('Day')
plt.ylabel('Transaction Count')
plt.grid(True)
plt.legend()
plt.show()

# Plot transaksi harian untuk bulan Oktober
plt.figure(figsize=(12, 6))
plt.plot(daily_transactions_october.index, daily_transactions_october.values, marker='o', color='orange', label='October')
plt.title('Daily Transactions in October')
plt.xlabel('Day')
plt.ylabel('Transaction Count')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Data untuk bulan September
september_data = df[(df['transaction_date'].dt.month == 9) & (df['transaction_date'].dt.year == 2024)]

# Data untuk bulan Oktober
october_data = df[(df['transaction_date'].dt.month == 10) & (df['transaction_date'].dt.year == 2024)]

# Kelompokkan data berdasarkan produk dan tanggal transaksi
september_grouped = september_data.groupby(['transaction_date', 'product_name']).size().unstack(fill_value=0)
october_grouped = october_data.groupby(['transaction_date', 'product_name']).size().unstack(fill_value=0)

# Visualisasikan data untuk melihat tren penurunan per produk di bulan September
plt.figure(figsize=(15, 10))
september_grouped.plot(marker='o', ax=plt.gca())
plt.title('Daily Transactions for Each Product in September')
plt.xlabel('Date')
plt.ylabel('Transaction Count')
plt.grid(True)
plt.legend(title='Product', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Visualisasikan data untuk melihat tren penurunan per produk di bulan Oktober
plt.figure(figsize=(15, 10))
october_grouped.plot(marker='o', ax=plt.gca())
plt.title('Daily Transactions for Each Product in October')
plt.xlabel('Date')
plt.ylabel('Transaction Count')
plt.grid(True)
plt.legend(title='Product', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Grafik ini menunjukkan perubahan jumlah transaksi untuk berbagai produk dari bulan September ke Oktober. Sumbu vertikal menampilkan perubahan jumlah transaksi, sedangkan sumbu horizontal menampilkan kategori produk.
Dari grafik ini, kita bisa melihat:
Penurunan transaksi terbesar terjadi pada produk "Keperluan Rumah Tangga" dan "Food & Drink", dengan penurunan sekitar 300 transaksi.
Peningkatan transaksi terbesar terjadi pada produk "Man Fashion" dan "Hotel", dengan peningkatan sekitar 200-300 transaksi.
Grafik ini membantu kita memahami produk mana yang mengalami peningkatan atau penurunan popularitas dari bulan September ke Oktober.

```python
# Analisis lebih lanjut untuk melihat produk mana yang mengalami penurunan terbesar
# Hitung total transaksi per produk di bulan September dan Oktober
september_totals = september_grouped.sum()
october_totals = october_grouped.sum()

# Gabungkan dan hitung selisih antara September dan Oktober
product_changes = pd.DataFrame({
    'September': september_totals,
    'October': october_totals
})
product_changes['Change'] = product_changes['October'] - product_changes['September']
product_changes = product_changes.sort_values('Change')

# Visualisasikan produk dengan penurunan terbesar
plt.figure(figsize=(15, 6))
product_changes['Change'].plot(kind='bar', color='red')
plt.title('Change in Transactions from September to October')
plt.xlabel('Product')
plt.ylabel('Change in Transaction Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Print the products with the largest decreases
print(product_changes.head(10))
```

```python
# Filter data untuk bulan September dan Oktober
df_september = df[df['transaction_date'].dt.month == 9]
df_october = df[df['transaction_date'].dt.month == 10]

# Produk yang mengalami penurunan
products_september = df_september['product_name'].value_counts()
products_october = df_october['product_name'].value_counts()

# Menghitung perubahan transaksi per produk
product_changes = products_october - products_september
product_changes = product_changes.sort_values()

# Produk dengan penurunan terbesar
top_declining_products = product_changes.head(5)
print("Top 5 Produk dengan Penurunan Terbesar:")
print(top_declining_products)

# Filter data untuk produk yang mengalami penurunan terbesar
declining_product_data = df[df['product_name'].isin(top_declining_products.index)]

# Grup berdasarkan produk dan tanggal untuk analisis korelasi
grouped_data = declining_product_data.groupby(['product_name', 'transaction_date']).agg({
    'user_id': 'count',
    'refund': 'sum',
    'discounts': 'sum'
}).reset_index()

# Pivot untuk memudahkan analisis korelasi
pivot_data = grouped_data.pivot(index='transaction_date', columns='product_name', values=['user_id', 'refund', 'discounts'])

# Analisis korelasi
for product in top_declining_products.index:
    product_data = pivot_data.xs(product, level=1, axis=1)
    print(f"Korelasi untuk {product}:")
    print(product_data.corr())

# Visualisasi perubahan transaksi
for product in top_declining_products.index:
    plt.figure(figsize=(12, 6))
    plt.plot(pivot_data.index, pivot_data['user_id'][product], label='Transactions')
    plt.plot(pivot_data.index, pivot_data['refund'][product], label='Refunds', linestyle='--')
    plt.plot(pivot_data.index, pivot_data['discounts'][product], label='Discounts', linestyle='-.')

    plt.title(f'Trend Harian untuk Produk: {product}')
    plt.xlabel('Tanggal')
    plt.ylabel('Jumlah')
    plt.legend()
    plt.show()
```

```python
# mengecek outlier di lis cols atau colom RFM menggunakan boxplot
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20,5))
for i, col in enumerate(cols):
    df_cluster.boxplot(column=col, ax=axes.flatten()[i], grid=False)

plt.show()
```

**Menghitung jumlah baris yang terindikasi outlier**

```python
# Menghitung Q1, Q3, dan IQR untuk setiap kolom
Q1 = df_cluster[cols].quantile(0.25)
Q3 = df_cluster[cols].quantile(0.75)
IQR = Q3 - Q1

# Menentukan batas bawah dan batas atas untuk setiap kolom
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Menentukan data yang bersih dan outlier untuk setiap kolom
cluster_clean = df_cluster[~((df_cluster[cols] < lower_bound) | (df_cluster[cols] > upper_bound)).any(axis=1)]
cluster_outlier = df_cluster[((df_cluster[cols] < lower_bound) | (df_cluster[cols] > upper_bound)).any(axis=1)]

print("Total baris data bersih:", len(cluster_clean))
print("Total baris data outlier:", len(cluster_outlier))
```

**Menghitung berapa persen outlier di kolom RFM**

```python
# Menghitung jumlah baris data bersih dan outlier
total_clean = len(cluster_clean)
total_outlier = len(cluster_outlier)

# Menghitung persentase outlier
percentage_outlier = (total_outlier / (total_clean + total_outlier)) * 100

print("Persentase data yang merupakan outlier: {:.2f}%".format(percentage_outlier))
```

**Menangani Outlier menggunakan metode IQR**

```python
# handling outliers
def handle_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Menentukan batas bawah dan batas atas
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Mengganti nilai outlier dengan nilai batas atas atau batas bawah terdekat
    df_cluster[column] = np.where(df_cluster[column] < lower_bound, lower_bound, df_cluster[column])
    df_cluster[column] = np.where(df_cluster[column] > upper_bound, upper_bound, df_cluster[column])

# Menggunakan fungsi handle_outliers_iqr untuk setiap kolom yang memiliki outlier
cols
for column in cols:
    handle_outliers_iqr(df, column)
```

```python
# normalisasi
cluster_scale = cluster_clean.copy()
scaler = MinMaxScaler()
cluster_scale[cols] = scaler.fit_transform(cluster_scale[cols])
cluster_scale
```

```python
# elbow
cluster_check = cluster_scale.copy()
cluster_check = cluster_scale.drop(['user_id'], axis=1)
cluster_check.head()
```

```python
distortions = []
K = range(2,11)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(cluster_check)
    distortions.append(kmeanModel.inertia_)

plt.figure(figsize=(7,4))
sns.set_style('white')
plt.plot(K, distortions,'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('The Elbow Showing the Optimal k')
plt.show()
```

```python
# Silhouette analysis
range_n_clusters = [2, 3, 4, 5, 6]

for num_clusters in range_n_clusters:
    # Inisialisasi model KMeans
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50, n_init=10)  # Menggunakan n_init secara eksplisit untuk menghilangkan peringatan FutureWarning
    kmeans.fit(cluster_check)

    # Mendapatkan label cluster untuk setiap data
    cluster_labels = kmeans.labels_

    # Menghitung silhouette score
    silhouette_avg = silhouette_score(cluster_check, cluster_labels)

    # Menampilkan hasil
    print("Untuk n_clusters={}, nilai silhouette score adalah {:.2f}".format(num_clusters, silhouette_avg))
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```
