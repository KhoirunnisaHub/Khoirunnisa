```python
#Converts bag-of-word vectors into arrays
import pandas as pd

index = ['text_kesehatan', 'text_olahraga', 'text_keuangan', 'query_text']
df = pd.DataFrame(data=bow_vectors, columns=feature_names, index=index)
df.head()
```

* index = ['text_sport', 'text_medical', 'text_finance', 'query_text']: Mendefinisikan sebuah list index yang berisi label untuk setiap teks dalam representasi BoW. Dalam hal ini, label-label tersebut adalah 'text_sport', * 'text_medical', 'text_finance', dan 'query_text'.
* df = pd.DataFrame(data=bow_vectors, columns=feature_names, index=index): Membuat sebuah dataframe Pandas (df) dengan menggunakan data dari bow_vectors, yang merupakan matriks BoW, dengan kolom-kolom yang diberi label menggunakan feature_names (nama fitur), dan baris-barisnya diberi label menggunakan index (label teks). Ini menghasilkan dataframe di mana setiap baris mewakili satu teks dan setiap kolom mewakili satu fitur (kata dalam BoW), dengan nilai-nilai yang menunjukkan frekuensi kemunculan setiap kata dalam setiap teks.
* df.head(): Mencetak lima baris pertama dari dataframe df untuk melihat bagaimana struktur datanya.

```python
# Normalize the BoW vectors
bow_texts_norm = []
for bow in bow_vectors:
  length = (sum(i*i for i in bow)) ** 0.5
  bow_norm = bow / length
  bow_texts_norm.append(bow_norm)

# Compute similarity using dot product
similarity_vector = []
bow_norm_query = bow_texts_norm[3]
for bow in bow_texts_norm[:3]:
  similarity_vector.append(sum(i*j for i,j in zip(bow,bow_norm_query)))
print(similarity_vector)

# Find the highest similarity
id_max_sim = similarity_vector.index(max(similarity_vector))
if (id_max_sim == 0):
  print ("The query text is classified as: kesehatan")
elif (id_max_sim == 1):
  print ("The query text is classified as: olahraga")
elif (id_max_sim == 2):
  print ("The query text is classified as: keungan")
```
