3.

4.
Update dataframe df dengan menambahkan hasil perhitungan soal nomor 2, beri nama kolom BMI.


```python
# Fungsi untuk menghitung BMI
def hitung_bmi(berat_badan, tinggi_badan):
    return berat_badan / ((tinggi_badan/100) ** 2)

# Menambahkan kolom BMI ke DataFrame
df['BMI'] = df.apply(lambda row: hitung_bmi(row['berat_badan'], row['tinggi_badan']), axis=1)
df
```


5. Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.

  - Normal: 18.5-24.9
  
  - Overweight: 25.0-29.9
  
  - Obesity: 30.0-34.9
  
  - Underweight: < 18.5





```python
# Fungsi untuk menentukan status BMI
def status_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi <= 24.9:
        return 'Normal'
    elif 25.0 <= bmi <= 29.9:
        return 'Overweight'
    elif 30.0 <= bmi <= 34.9:
        return 'Obesity'
    else:
        return 'Extreme Obesity'


# Menambahkan kolom Status BMI ke DataFrame
df['Status BMI'] = df['BMI'].apply(status_bmi)

# Menampilkan DataFrame setelah update
df
```

**SOAL 2**

Perusahaan Investasi Jamin-Sugih Co kami memberikan penawaran investasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun.

Buatlah program Python yang dapat menerima input dari pengguna dalam bentuk:
Modal Investasi (bilangan bulat antara 0-1.000.000.000)
Jangka Waktu Investasi (bilangan bulat antara 1-25 dalam tahun)
Selain itu, program akan menampilkan error message bahwa modal atau tenure tidak sesuai ketentuan.

Program akan menampilkan informasi akumulasi investasi setiap tahun, menggunakan iterasi untuk setiap tahun.

Contoh:

Input

>> Money Capital : 100000

>> Investment Tenure (years) : 2

Output
>> Year 1, your total investment : 125000.0

>> Year 2, your total investment : 156250.0


```python
def calculate_investment(capital, tenure):
    if 0 < capital <= 1000000000 and 0 < tenure <= 25:
        annual_profit_percentage = 0.25

        for year in range(1, tenure + 1):
            capital *= (1 + annual_profit_percentage)
            print(f"Year {year}, your total investment: {capital:.2f}")

    else:
        print("Error: Input modal atau tenure tidak sesuai ketentuan.")

try:
    capital = int(input("Money Capital (0-1,000,000,000): "))
    tenure = int(input("Investment Tenure (1-25 years): "))
    calculate_investment(capital, tenure)

except ValueError:
    print("Error: Masukkan bilangan bulat yang valid.")
```
